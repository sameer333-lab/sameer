# Exploratory Data Analysis (EDA)
Heart Disease Rates

```python
# Calculate the prevalence of heart disease
heart_disease_rate = df['target'].mean() * 100
# Assuming 'target' is the column indicating heart disease
print(f'Heart Disease Rate: {heart_disease_rate:.2f}%')
```

```python
# Assuming 'target' is 1 for heart disease and 0 for no heart disease
if 'target' not in df.columns:
    raise ValueError("The dataset must have a 'target' column indicating heart disease presence.")
```

```python
# Assuming the 'gender' column name is 'sex'
gender_column_name = 'sex'  # Replace with the correct column name if different
print(df[gender_column_name].unique())
```

```python
#'gender' is a categorical variable
# Assuming 'sex' is encoded as 1 for male and 0 for female,
df[gender_column_name] = df[gender_column_name].map({1: 'Male', 0: 'Female'})
```

```python
df[gender_column_name]
```

```python
# Now, Calculate the heart disease rate by gender
gender_heart_disease = df.groupby(gender_column_name)['target'].mean() * 100
print(gender_heart_disease)
```

# Visualization

```python
sns.countplot(data=df, x=gender_column_name, hue='target')
plt.title('Heart Disease by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

# Visualization: Age Distribution

```python
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=20, edgecolor='k')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# Visualization: Heart Disease by Age Group

```python
df['age_group'] = pd.cut(df['age'], bins=[20, 30, 40, 50, 60, 70, 80])
age_heart_disease = df.groupby('age_group')['target'].mean() * 100
print(age_heart_disease)

age_heart_disease.plot(kind='bar', figsize=(10, 6))
plt.title('Heart Disease Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Heart Disease Rate (%)')
plt.show()
```
